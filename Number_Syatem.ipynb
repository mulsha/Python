```python
binary = input("Enter binary Number: ")

print("Binary to Decimal :: ",int(binary,2))
```

```python
decimal = int(input("Enter decimal Number: "))
print("Decimal to Binary :: ",bin(decimal))
```

```python
decimal = int(input("Enter decimal Number: "))
print("Decimal to Octal :: ",oct(decimal))
```

```python
decimal = int(input("Enter decimal Number: "))
print("Decimal to Hexa :: ",hex(decimal))
```

```python
octal = input("Enter Octal Number: ")
print("Octal to Decimal :: ",int(octal, 8))
```

```python
Hexa = input("Enter Hexa Number: ")
print("Hexa to Decimal :: ",int(Hexa, 16))
```

```python
unicode = int(input("Enter Unicode Number : "))
print("Unicode to Character :: ",chr(unicode))
```

```python
character = input("Enter Character Number: ")
print("Character to Unicode :: ",ord(character))
```
